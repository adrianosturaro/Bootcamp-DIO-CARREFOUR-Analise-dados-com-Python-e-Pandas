# Trabalhando com datas

## Importando a Biblioteca

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Fazendo a leitura dos arquivos

```python
df1 = pd.read_excel(r".\datasets\Aracaju.xlsx")
df2 = pd.read_excel(r".\datasets\Fortaleza.xlsx")
df3 = pd.read_excel(r".\datasets\Natal.xlsx")
df4 = pd.read_excel(r".\datasets\Recife.xlsx")
df5 = pd.read_excel(r".\datasets\Salvador.xlsx")
```

## Agrupando os arquivos

```python
df = pd.concat([df1, df2, df3, df4, df5])
```

## Criando a coluna Receitas

```python
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

## Criando nova coluna com o ano

```python
df['Ano'] = df['Data'].dt.year
df.sample(8)
```

## Criando coluna com mês e dia

```python
df['Mes'], df['Dia'] = df['Data'].dt.month, df['Data'].dt.day
df.sample(5)
```

## Criando a coluna Trimestre

```python
df['Trimestre'] = df['Data'].dt.quarter
df.sample(5)
```

## Número de Vendas por loja

### Dados

```python
df['LojaID'].value_counts(ascending=False)
```

### Gráfico

```python
df['LojaID'].value_counts(ascending=False).plot.bar()
```

```python
df['LojaID'].value_counts(ascending=True).plot.barh();
```

## Gráfico pizza receita por ano

```python
df.groupby([df['Ano']])['Receita'].sum().plot.pie();
```

## Vendas por cidade

```python
df['Cidade'].value_counts()
```

### Adicionando um título a modificando o título dos eixos

```python
df['Cidade'].value_counts().plot.bar(title='Total de vendas por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas');
```

### Modificando a cor

```python
df['Cidade'].value_counts().plot.bar(title='Total de vendas por Cidade', color='red');
```

### Alterando o estilo

```python
plt.style.use('ggplot')
df['Cidade'].value_counts().plot.bar(title='Total de vendas por Cidade', color='red'); 
```

## Venda por mes Gráfico de linha

```python
df.groupby(df['Mes'])['Qtde'].sum().plot(title='Total de Produtos Vendidos por Mês')
plt.xlabel('Mês')
plt.ylabel('Total de Produtos Vendidos');
plt.legend();
```

## Vendas por mês no ano do 2019

```python
df_2019 = df[df['Ano']==2019]
df_2019.groupby(df_2019['Mes'])['Qtde'].sum().plot(title='Total de Produtos Vendidos por Mês em 2019', marker='v')
plt.xlabel('Mês')
plt.ylabel('Total de Produtos Vendidos');
plt.legend();
```

## Histograma

```python
plt.hist(df['Qtde'], color='orangered');
```

## Gráfico de disperssão

```python
plt.scatter(x=df_2019['Dia'], y = df_2019['Receita']);
```

## Salvando em PNG

```python
df_2019.groupby(df_2019['Mes'])['Qtde'].sum().plot(marker='v')
plt.title('Total de Produtos Vendidos por Mês em 2019')
plt.xlabel('Mês')
plt.ylabel('Total de Produtos Vendidos');
plt.legend()
plt.savefig('Grafico QTD x Mes')
```
